```python

import pandas as pd
import matplotlib.pyplot as plt

# Input data
data = {
    'Year': [2018, 2019, 2020, 2021, 2022],
    'Number of persons (x 1,000)': [2717, 2747, 2800, 2847, 2885],
    'Number with income (x 1,000)': [2129, 2173, 2126, 2229, 2332],
    'Aggregate income (x $1,000,000)': [144016, 151166, 145999, 158297, 171744],
    'Average income (excluding zeros)': [67700, 69600, 68700, 71000, 73700],
    'Median income (excluding zeros)': [54900, 55900, 57400, 56300, 57000]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display DataFrame
print(df)

# Plotting the data
plt.figure(figsize=(10, 6))

# Number of persons over the years
plt.subplot(2, 2, 1)
plt.plot(df['Year'], df['Number of persons (x 1,000)'], marker='o')
plt.title('Number of Persons')
plt.xlabel('Year')
plt.ylabel('Number (x 1,000)')

# Number with income over the years
plt.subplot(2, 2, 2)
plt.plot(df['Year'], df['Number with income (x 1,000)'], marker='o')
plt.title('Number with Income')
plt.xlabel('Year')
plt.ylabel('Number (x 1,000)')

# Aggregate income over the years
plt.subplot(2, 2, 3)
plt.plot(df['Year'], df['Aggregate income (x $1,000,000)'], marker='o')
plt.title('Aggregate Income')
plt.xlabel('Year')
plt.ylabel('Income (x $1,000,000)')

# Average and Median income over the years
plt.subplot(2, 2, 4)
plt.plot(df['Year'], df['Average income (excluding zeros)'], marker='o', label='Average Income')
plt.plot(df['Year'], df['Median income (excluding zeros)'], marker='o', label='Median Income')
plt.title('Average and Median Income')
plt.xlabel('Year')
plt.ylabel('Income ($)')
plt.legend()

plt.tight_layout()
plt.show()



```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster

# Input data
data = {
    'Year': [2018, 2019, 2020, 2021, 2022],
    'Number of persons (x 1,000)': [2717, 2747, 2800, 2847, 2885],
    'Number with income (x 1,000)': [2129, 2173, 2126, 2229, 2332],
    'Aggregate income (x $1,000,000)': [144016, 151166, 145999, 158297, 171744],
    'Average income (excluding zeros)': [67700, 69600, 68700, 71000, 73700],
    'Median income (excluding zeros)': [54900, 55900, 57400, 56300, 57000]
}

# Create DataFrame
df = pd.DataFrame(data)

# Data Preprocessing
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df.drop('Year', axis=1))

# K-means Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(scaled_data)
df['KMeans_Cluster'] = kmeans_labels

# DBSCAN Clustering
dbscan = DBSCAN(eps=0.5, min_samples=2)
dbscan_labels = dbscan.fit_predict(scaled_data)
df['DBSCAN_Cluster'] = dbscan_labels

# Hierarchical Clustering
linked = linkage(scaled_data, method='ward')
hierarchical_labels = fcluster(linked, t=3, criterion='maxclust')
df['Hierarchical_Cluster'] = hierarchical_labels

# Visualization
plt.figure(figsize=(14, 8))

# K-means Clustering
plt.subplot(2, 2, 1)
sns.scatterplot(x='Number of persons (x 1,000)', y='Aggregate income (x $1,000,000)', hue='KMeans_Cluster', data=df, palette='viridis')
plt.title('K-means Clustering')

# DBSCAN Clustering
plt.subplot(2, 2, 2)
sns.scatterplot(x='Number of persons (x 1,000)', y='Aggregate income (x $1,000,000)', hue='DBSCAN_Cluster', data=df, palette='viridis')
plt.title('DBSCAN Clustering')

# Hierarchical Clustering
plt.subplot(2, 2, 3)
sns.scatterplot(x='Number of persons (x 1,000)', y='Aggregate income (x $1,000,000)', hue='Hierarchical_Cluster', data=df, palette='viridis')
plt.title('Hierarchical Clustering')

# Dendrogram for Hierarchical Clustering
plt.subplot(2, 2, 4)
dendrogram(linked, labels=df['Year'].values, leaf_rotation=90)
plt.title('Hierarchical Clustering Dendrogram')

plt.tight_layout()
plt.show()

# Display DataFrame with cluster labels
print(df)
```
